```python
import matplotlib.pyplot as plt
import numpy as np

# Provided timing data
# timing_data = {
#     "8 Processes (T1)": [12.06, 12.44,12.69,14.29,14.95],
#     "16 Processes (T2)": [11.89, 12.03,12.36,12.27,12.33],
#     "32 Processes (T3)": [14.19, 17.20,17.59,19.50,23.78,35.49]
# }


timing_data = {
    "8 Processes (T1)": [12.06, 12.44],
    "16 Processes (T2)": [11.89, 12.03],
    "32 Processes (T3)": [14.19, 17.20]
}

# Extract labels, means, and standard deviations
labels = list(timing_data.keys())
means = [np.mean(times) for times in timing_data.values()]
std_devs = [np.std(times) for times in timing_data.values()]

# Set up the 2x2 plot grid
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Execution Time Analysis Across Different Process Counts on Dataset 1", fontsize=16)



# 2. Scatter Plot with Means and Individual Time Labels
# Scatter Plot with Means and Individual Time Labels
for i, (label, times) in enumerate(timing_data.items()):
    # Scatter individual times
    axs[0, 0].scatter([label] * len(times), times, color='salmon', s=80)
    # Label individual times
    # for time in times:
    #     axs[0, 0].text(label, time + 0.1, f'{time:.2f}', ha='center', color='darkred')
    # Plot mean point for each label
    axs[0, 0].plot(label, means[i], color='blue', marker='o', markersize=8)
    # Label mean
    axs[0, 0].text(label, means[i] + 0.1, f'{means[i]:.2f}', ha='center', color='blue')

# Add labels only once for legend
axs[0, 0].scatter([], [], color='salmon', s=80, label='Individual times')  # Empty scatter for legend
axs[0, 0].plot([], [], color='blue', marker='o', linestyle='-', label='Mean time')  # Empty plot for legend

axs[0, 0].set_ylabel('Execution Time (s)')
axs[0, 0].set_title('Scatter Plot of Execution Times with Means')
axs[0, 0].legend()  # Add legend


# 3. Line Plot for Timing Trends with Mean Labels
axs[0, 1].plot(labels, means, color='purple', marker='o', linestyle='-', linewidth=2, markersize=8)
axs[0, 1].fill_between(labels, np.array(means) - np.array(std_devs), np.array(means) + np.array(std_devs), color='lavender', alpha=0.5)
for i, mean in enumerate(means):
    axs[0, 1].text(labels[i], mean + 0.1, f'{mean:.2f}', ha='center', color='purple')
axs[0, 1].set_ylabel('Execution Time (s)')
axs[0, 1].set_title('Line Plot for Timing Trends')


# 1. Bar Plot with Error Bars and Mean Labels
# Bar Plot with Error Bars and Mean Labels at Center of Error Bar
axs[1, 0].bar(labels, means, yerr=std_devs, capsize=5, color='skyblue')
for i, mean in enumerate(means):
    # Position the mean label at the center of the error bar
    y_position = mean + (std_devs[i] / 4)
    axs[1, 0].text(i, y_position, f'{mean:.2f}', va='center', ha='left', color='blue')
axs[1, 0].set_ylabel('Execution Time (s)')
axs[1, 0].set_title('Bar Plot with Error Bars')


# 4. Box-Quartile Plot with Individual Points
axs[1, 1].boxplot(timing_data.values(), vert=True, patch_artist=True, labels=labels)
for i, (label, times) in enumerate(timing_data.items()):
    # Add individual points on the box plot for clarity
    y = times
    x = np.random.normal(i + 1, 0.04, size=len(y))  # Slightly jitter points for visibility
    # axs[1, 1].scatter(x, y, color='salmon', alpha=0.7)
    for j, time in enumerate(times):
        axs[1, 1].text(i + 1, time + 0.1, f'{time:.2f}', ha='center', color='darkred')
axs[1, 1].set_ylabel('Execution Time (s)')
axs[1, 1].set_title('Box-Quartile Plot of Execution Times')

# Adjust layout and show the plots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Add space for the main title
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Provided timing data
timing_data = {
    "8 Processes (T1)": [12.06, 12.44,12.69,14.29,14.95],
    "16 Processes (T2)": [11.89, 12.03,12.36,12.27,12.33],
    "32 Processes (T3)": [14.19, 17.20,17.59,19.50,23.78,35.49]
}

# Extract labels and data
labels = list(timing_data.keys())
data = list(timing_data.values())

# Create the boxplot
fig, ax = plt.subplots()
box = ax.boxplot(data, labels=labels, showmeans=True)

# Set title and labels
ax.set_title("Dataset 1")
ax.set_xlabel("Process Count per Test Case")
ax.set_ylabel("Execution Time (seconds)")

# Annotate min, max, Q1, Q3, median, and mean for each boxplot
for i in range(len(data)):
    min_val = box['caps'][i * 2].get_ydata()[0]           # Minimum value
    max_val = box['caps'][i * 2 + 1].get_ydata()[0]       # Maximum value
    q1 = box['boxes'][i].get_ydata()[0]                   # First quartile (Q1)
    median_val = box['medians'][i].get_ydata()[0]         # Median
    q3 = box['boxes'][i].get_ydata()[2]                   # Third quartile (Q3)
    mean_val = box['means'][i].get_ydata()[0]             # Mean value

    # Annotate min and max values
    ax.text(i + 1, max_val, f"{max_val:.2f}", ha='center', va='bottom', color='blue')
    ax.text(i + 1, min_val, f"{min_val:.2f}", ha='center', va='top', color='blue')

    # Annotate Q1, median, Q3, and mean values
    # ax.text(i + 1, q1, f"Q1: {q1:.2f}", ha='left', va='top', color='purple')
    ax.text(i + 1, median_val, f" {median_val:.2f}", ha='center', va='bottom', color='green', fontweight='bold')
    # ax.text(i + 1, q3, f"Q3: {q3:.2f}", ha='left', va='bottom', color='purple')
    # ax.text(i + 1, mean_val, f"Mean: {mean_val:.2f}", ha='center', va='bottom', color='orange', fontstyle='italic')

plt.show()
```
